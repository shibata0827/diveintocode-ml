```python
import h5py
f = h5py.File('./../../output/models/IMAGENETTE/ResNet18_TD.h5', 'r')
print(list(f.keys()))
# will get a list of layer names which you can use as index
d = f['model_weights']#['dense_1']['kernel:0']
# # <HDF5 dataset "kernel:0": shape (128, 1), type "<f4">
type(d.values)
# d[0] == array([-0.14390108], dtype=float32)
```

```python
def PrintOnlyDataset(name, obj):
    if isinstance(obj, h5py.Dataset):
        print(name)
        # print('\t',obj)

with h5py.File('./../../output/models/IMAGENETTE/ResNet18_TD.h5','r') as f:
    f.visititems(PrintOnlyDataset)
    print(f['res2a_branch1/res2a_branch1_W_1:0'].value)    
    print(f['res2a_branch1/res2a_branch1_W_1:0'].value.shape)
```
