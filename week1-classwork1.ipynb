【問題1】100日目の米粒の数

100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

```python
import matplotlib.pyplot as plt
%matplotlib inline

#初期値入力
num = 1
list_num = []
sum_num = num
list_sum = []

#リスト化
list_num.append(num)
list_sum.append(num)

#米を貰う行動の繰り返し
#初期値入力で1回行なっているのでforは99回
for i in range(100-1):
    num = num * 2
    list_num.append(num) # 貰う米粒数のリスト追加
    sum_num += num # 今までの合計を算出
    list_sum.append(sum_num) #合計のリスト追加

#日毎のグラフ出力
plt.title('kome nums')
plt.xlabel('day')
plt.ylabel('num')
plt.plot(list_num);
```

```python
#累計のグラフ出力
plt.title('kome nums')
plt.xlabel('day')
plt.ylabel('sum_num')
plt.plot(list_sum);
```

【問題2】100日目以外の米粒の数

この逸話は30日だったり81日だったりと日数が曖昧です。どの話にも対応できるように日数を引数とした 関数 を作成しましょう。関数を作成した上で、コースを受講する日数の間でいくつ米粒をもらえるかを答えてください。前の問題同様に折れ線グラフも出力してください。

以下の雛形には関数としての枠組みだけではなく、引数や返り値についてのコメントも記載してあります。参考にして、自分なりにも分かりやすくコメントを書き換えてください。

```python
def compute_sorori_shinzaemon(day=100):
    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    Parameteres
    --------------
    day: int
        米を貰う日数 (default : 100)

    Returns
    --------------
    list_num : list
        ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
    list_sum : list
        ある日までに累計で貰う米の数のリスト（リストの長さはdayに等しくなる）
    """
    
    #初期値入力
    num = 1
    list_num = []
    sum_num = num
    list_sum = []

    #リスト化
    list_num.append(num)
    list_sum.append(num)

    #米を貰う行動の繰り返し
    #初期値入力で1回行なっているのでfor内では−１回
    for i in range(day-1):
        num = num * 2
        list_num.append(num) # 貰う米粒数のリスト追加
        sum_num += num # 今までの合計を算出
        list_sum.append(sum_num) #合計のリスト追加

    return list_num, list_sum #貰う米粒数のリスト、合計のリストを返す

#10日までで計算
list_num2, list_sum2 = compute_sorori_shinzaemon(day=10)
```

```python
#10日までに貰う数　日毎グラフ
plt.title('kome nums')
plt.xlabel('day')
plt.ylabel('num')
plt.plot(list_num2);
```

```python
#10日までに貰う数　累計グラフ
plt.title('kome sum_num')
plt.xlabel('day')
plt.ylabel('sum_num')
plt.plot(list_sum2);
```

【問題3】何人が何日生活できる米か

この米があれば何人が何日生活できるか計算する関数を作成してください。コースを受講する期間に貰える米粒で、このクラスにいる人数であれば、何日間生活できるかを出力してみましょう。

関数の最低限の入力は 米粒の数(int) と 人数(int) 、出力は 生活できる日数(int) です。関数名やコメントなどは分かりやすく考えて作成してください。

ここでは1日に必要な米粒の数は提示しません。自分なりに調べて、計算式に落とし込んでください。どう調べ、何を根拠にしたかもコメントを残しておいてください。

```python
#成人男性の最低限カロリー(kcal / 日) ： 厚生労働省「日本人の食事摂取基準」
men_cal = 2000

#米一粒の重さ(g / 粒)：新潟大学　お米Q&A
kome_g = 0.02

#米のカロリー(kcal / g)：グラムわかる写真館
kome_cal = 3.56


def survive(kome_num, people_num=14):
    #DICメンバー人数：富永さん含む
    
    
    #米の総カロリー
    total_kome_cal = kome_num * kome_g * kome_cal
    
    #1日当たりの全員分の最低限必要カロリー
    total_man_cal = people_num * men_cal
    
    #米のみで生活できる日数（四捨五入）
    survive_day = total_kome_cal // total_man_cal
    print('コース期間113日で貰った米で生き残れる日数：{}日'.format(survive_day))
    print('年数に換算すると：{}年'.format(survive_day//365))


#コース受講期間で貰えるコメの数を、問題２の関数から収得する
#コース受講期間 113日（1月9日から4月30日）
list_num3, list_sum3 = compute_sorori_shinzaemon(113)
kome_num = list_sum3[-1]

#受講期間で貰える米の数を「survive関数」に代入
#人数は初期値（14人のまま）
survive(kome_num)
```
