# 【問題5】Pythonの関数化
問題1から問題3までの範囲をまとめて行うPythonの関数を雛形を参考に作成してください。

そして、以下の3式の勾配を0.1間隔で求めてndarrayで出力してください。

また、問題4と同様にグラフを描いてください。

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    変化量を使い勾配を計算する。

    Parameters
    ----------------
    function : function
    求めたい関数、xのndarrayを入れるとyのndarrayを返すもの。
    x_range : tuple
    np.arange()の引数と同じ方法で範囲を指定する。

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
    xとyを結合したもの。nはx_rangeに依存する。
    gradient : ndarray, shape(n-1,)
    関数の勾配。差をとるため要素は1つ減る
    """

    x = np.arange(*x_range)
    y = function(x)
    
    array_xy = np.vstack((x, y)).T
    sa = array_xy[1:] - array_xy[:-1]
    gradient = sa[:, 1] / sa[:, 0]
    return array_xy, gradient


# 関数の中のxをarray_xとする
def function1(array_x):
    array_y1 = 1/2*array_x + 1
    return array_y1


def function2(array_x):
    array_y2 = array_x**2
    return array_y2


def function3(array_x):
    array_y3 = 2*array_x**2 + 2**array_x
    return array_y3


def function4(array_x):
    array_y4 = np.sin(array_x**(1/2))
    return array_y4
```

```python
# y=1/2x+1
array_xy, gradient = compute_gradient(function1)
print("array_xy\n{}\n".format(array_xy))
print("gradient(最初の５つ)\n{}".format(gradient[:5]))

plt.title('function1')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[:, 0], array_xy[:, 1])
plt.show()

plt.title('function1:grad')
plt.xlabel('x')
plt.ylabel('y/x')
plt.plot(array_xy[1:, 0], gradient)
plt.show()
```

```python
# y=x^2
array_xy, gradient = compute_gradient(function2)
print("array_xy\n{}\n".format(array_xy))
print("gradient(最初の５つ)\n{}".format(gradient[:5]))

plt.title('function2')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[:, 0], array_xy[:, 1])
plt.show()

plt.title('function2:grad')
plt.xlabel('x')
plt.ylabel('y/x')
plt.plot(array_xy[1:, 0], gradient)
plt.show()
```

```python
# y=2x^2 + 2^x
array_xy, gradient = compute_gradient(function3)
print("array_xy\n{}\n".format(array_xy))
print("gradient(最初の５つ)\n{}".format(gradient[:5]))

plt.title('function3')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[:, 0], array_xy[:, 1])
plt.show()

plt.title('function3:grad')
plt.xlabel('x')
plt.ylabel('y/x')
plt.plot(array_xy[1:, 0], gradient)
plt.show()
```

```python
# y=sin(x^1/2)、(0 <= x <= 50)
array_xy, gradient = compute_gradient(function4, x_range=(0, 50.1, 0.1))
print("array_xy\n{}\n".format(array_xy))
print("gradient最初の５つ)\n{}".format(gradient[:5]))

plt.title('function4')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[:, 0], array_xy[:, 1])
plt.show()

plt.title('function4:grad')
plt.xlabel('x')
plt.ylabel('y/x')
plt.plot(array_xy[1:, 0], gradient)
plt.show()
```

# 【問題6】最小値を求める
問題5で扱った関数のyの最小値を求めます。xの範囲も問題5と同様とします。

今回は勾配降下法ではなく、ndarrayのminメソッド、argminメソッドを用いてください。minメソッドは最小値を、argminメソッドは最小値のインデックスを取得することができます。

また、最小のyをとる前後の勾配を表示してください。

```python
# y=1/2x+1
array_xy, gradient = compute_gradient(function1)
print("最小のインデックス：{}、値：{}".format(
    array_xy[:, 1].argmin(), array_xy[:, 1].min()
    ))
print("最小値の前後の勾配：前{}、後{}".format(
    gradient[array_xy[:, 1].argmin()-1], gradient[array_xy[:, 1].argmin()+1]
    ))
```

```python
# y=2x^2 + 2^x
array_xy, gradient = compute_gradient(function3)
print("最小のインデックス：{}、値：{}".format(
    array_xy[:, 1].argmin(), array_xy[:, 1].min()
    ))
print("最小値の前後の勾配：前{}、後{}".format(
    gradient[array_xy[:, 1].argmin()-1], gradient[array_xy[:, 1].argmin()+1]
    ))
```
